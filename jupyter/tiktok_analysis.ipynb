```python
import pandas as pd
df = pd.read_excel(r'C:\Users\sumey\Desktop\temer realstate\analysis_1\social_media_analysis_pipline\files\final_data_with_reel.xlsx')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Importing seaborn for a nice color palette

# List of company names
company_names = ['TEMER Properties', 'Palm Real Estate', 'Metropolitan Real Estate', 'Amibara Real Estate',
                 'Noah Real Estate', 'Reality Real Estate', 'Ovid Real Estate', 'DMC Real Estate',
                 'Demahope Real Estate', 'Hosea Real Estate', 'Akoya Properties', 'GIFT Real Estate',
                 'Janbo Real Estate', 'Horra Real Estate', 'Cosmopolitan Real Estate','Jenboro Real Estate']

# Generating a color palette with a distinct color for each company
colors = plt.cm.get_cmap('viridis', len(company_names))

# Loop through each company for filtering and plotting
for idx, company in enumerate(company_names):
    # Filter data for the current company
    company_data = df[df['company_name'] == company]
    
    # Create a pivot table for the filtered data
    pivot_data = company_data.pivot_table(index='week_number', values='tiktok_posts_engagement_total' )

    
    # Plotting with different colors
    fig, ax = plt.subplots(figsize=(10, 5))
    pivot_data.plot(kind='bar', color=[colors(idx)], logy=True, ax=ax)  # Applying logarithmic scale
    plt.title(f'tiktok Total Engagement per Week for {company}')
    plt.xlabel('Week Number')
    plt.ylabel('Total tiktok Engagement (log scale)')
    plt.legend([company])
    plt.tight_layout()  # Adjust layout to fit everything comfortably
     
    plt.show()
plt.tight_layout() 
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df has been filtered and processed as you described

# Determine unique week numbers
week_numbers = df['week_number'].unique()

# Plotting one graph per week number
for week in week_numbers:
    week_data = df[df['week_number'] == week]
    week_data = week_data.sort_values(by='company_name')

    fig, ax = plt.subplots(figsize=(12, 8))
    colors = plt.cm.viridis(np.linspace(0, 1, len(week_data)))  # color mapping
    bars = ax.bar(week_data['company_name'], week_data['tiktok_posts_engagement_total'], color=colors)
    plt.title(f'Number of TikTok engagments in Week {week}')
    plt.xlabel('Company Name')
    plt.ylabel('Number of TikTok engagments')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Annotate number of views or special condition on top of each bar
    for bar, (posts, engagement) in zip(bars, week_data[[ 'tiktok_posts_engagement_total','tiktok_posts']].itertuples(index=False)):
        label = f'{engagement}'
        # Adjust the x position slightly if the alignment is off
        ax.annotate(label,
                    xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')  # Ensure horizontal alignment is centered

    plt.show()
```
